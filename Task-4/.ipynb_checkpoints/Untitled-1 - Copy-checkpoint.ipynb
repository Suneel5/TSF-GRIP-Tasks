```python
def wordcloud(column_name):
    # make wordcloud from the text of given column
    text=" ".join(df[df[column_name].notna()][column_name])
    wordcloud = WordCloud(width=800, height=500, max_font_size=200, background_color='white').generate(text)
    plt.figure(figsize=(10, 10))
    plt.imshow(wordcloud, interpolation="bilinear")
    plt.axis("off")
    plt.show()
```

# No of terrorist attack and death by terrorist attacks on dirrerent years

```python
yearwise_df=pd.DataFrame(df['iyear'].value_counts())
yearwise_df.columns=['Nattacks']
yearwise_df.sort_index(inplace=True)
yearwise_df.head()
```

```python
yearwise_death=df.groupby('iyear').sum()['nkill']
yearwise_death.head()
```

```python
plt.figure(figsize=(10,7))
plt.plot(yearwise_df.index,yearwise_df.iloc[:, 0],label='No of Terrorist attacks')
plt.plot(yearwise_death.index,yearwise_death.values,label='No of death ')
plt.xlabel('Year')
plt.title('No of terrorist attack and death by terrorist attacks on different years')
plt.legend()
plt.show()
```

# Death caused by terrorist attacks in different years

```python
fig=plt.figure(figsize=(18,8))
sns.barplot(yearwise_death.index,yearwise_death.values)
# sns.barplot(yearwise_df.index,yearwise_df.iloc[:, 0])
# for index,value in enumerate(yearwise_death.values,1970):
#     plt.text(index, value, str(value))
plt.ylabel('No of Deaths by  Terrorist attack')
plt.xlabel('Years')
plt.title('No of  Death Caused by terrorist attacks on different years')
fig.autofmt_xdate()
plt.show()
```

# No of terrorist attack vs Death caused by It in each year

```python
plt.scatter(yearwise_df.iloc[:, 0],yearwise_death.values)
plt.xlabel('No of Terrorist Attack')
plt.ylabel('No of deaths ')
plt.title(' No of terrorist attack vs Death caused by Terrorist attack in each year')
```

#  Which countries are mostly victim of Terrorist attack?

```python
wordcloud('country_txt')
```

```python
countrywise=df['country_txt'].value_counts().head(10)
```

```python
fig=plt.figure(figsize=(13,10))
sns.barplot(countrywise.index,countrywise.values)
for index, value in enumerate(countrywise.values):
    plt.text(index-0.3,value, str(value))
plt.ylabel('Country')
plt.xlabel('No of Terrorist attack')
plt.title('Countries that are mostly victim of Terrorist attack(Top 10)')
fig.autofmt_xdate()
plt.show()
```

# which group of terrorist used to cause such incident most?

```python
namewise=df.drop(df[df['gname']=='Unknown'].index)['gname'].value_counts().head(10)
```

```python
plt.figure(figsize=(13,8))
sns.barplot(namewise.values,namewise.index)
for index, value in enumerate(namewise.values):
    plt.text(value, index, str(value))
plt.ylabel('Name of terrorist group')
plt.xlabel('No of Terrorist attack')
plt.title(' which group of terrorist used to cause such incident most?(Top10)')
plt.show()
```

#  Type of weapen that are Mostly used in Terrorist attack!

```python
weapentype_wise=df['weaptype1_txt'].value_counts().head(6)
```

```python
plt.figure(figsize=(13,13))
squarify.plot(sizes=weapentype_wise.values,label=weapentype_wise.index,alpha=0.9)
plt.axis('off')
plt.title("Type of weapen that are Mostly used in Terrorist attack!")
plt.show()
```

# Total numbers of  Deaths caused by different group of terrorist group

```python
gname_kill=df.drop(df[df['gname']=='Unknown'].index).groupby(by='gname')['nkill'].sum()
gname_kill=gname_kill.sort_values(ascending=False).head(10)
```

```python
plt.figure(figsize=(13,8))
sns.barplot(gname_kill.values,gname_kill.index)
for index, value in enumerate(gname_kill.values):
    plt.text(value, index, str(value))
plt.ylabel('Name of terrorist group')
plt.xlabel('No of Deaths')
plt.title('Total numbers of  Deaths caused by different group of terrorist group(Top10)')
plt.show()
```

# Total No of deaths caused based on weapon type used!

```python
weapon_kill=df.groupby(by='weaptype1_txt')['nkill'].sum()
weapon_kill=weapon_kill.sort_values(ascending=False).head(5)
```

```python
plt.figure(figsize=(13,8))
sns.barplot(weapon_kill.index,weapon_kill.values)
# for index, value in enumerate(weapon_kill.values):
#     plt.text(value, index, str(value))
plt.ylabel('Name of Weapon type')
plt.xlabel('No of Deaths')
plt.title('Total No of deaths caused based on weapon type used!')
plt.show()
```

```python
plt.figure(figsize=(12,12))
squarify.plot(sizes=weapon_kill.values,label=weapon_kill.index,alpha=.9)
plt.axis('off')
plt.title('Total No of deaths caused based on weapon type used!')
plt.show()
```

# Total No of deaths caused based on attack type
